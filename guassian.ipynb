```python
esl= pd.read_csv("esl.csv")
esl  = esl[["DBN", "Mean Scale Score", "Level 3 %", "Level 4 %"]]
esldata = agg_dataset()
esldata.load_seller(esl, ["DBN"], "esl",False)

# number of attributes
m = len(esldata.X)
Xrange = esldata.data[esldata.X].max()-esldata.data[esldata.X].min()
# esldata.data[esldata.X] = (esldata.data[esldata.X] - esldata.data[esldata.X].min() - Xrange/2)/(Xrange/2) 
esldata.data[esldata.X] = (esldata.data[esldata.X] - esldata.data[esldata.X].min() - Xrange/2)/(Xrange/2 * (m)**0.5) 
```

```python
esldata.data
```

```python
square_sum = None
for i in range(len(esldata.X)):
    col = esldata.X[i]
    if i == 0:
        square_sum = np.square(esldata.data[col])
    else:    
        square_sum += np.square(esldata.data[col])
```

```python
# just to double check max is 1 and normalize is correct
min(square_sum)
```
