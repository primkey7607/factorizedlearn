```python
import pandas as pd
```

```python
gender = pd.read_csv("gender.csv")
math = pd.read_csv("math.csv")
```

```python
%timeit pd.merge(gender,math,how='inner', on="DBN")
```

```python
gender.set_index('DBN', inplace=True)
math.set_index('DBN', inplace=True)
```

```python
%timeit pd.merge(gender,math,how='inner', on="DBN")
```

```python
%timeit pd.merge(gender,math,how='inner', left_index=True, right_index=True)
```

```python
pd.merge(gender,math,how='inner', left_index=True, right_index=True)
```

Conclusion: index join speed up by 2x!!!

```python
%timeit gender.join(math, on="DBN", how='inner', lsuffix='_x', rsuffix='_y')
```
